```python
import pandas as pd
```

```python
df = pd.read_csv("reviews_booking.csv") 
df.head()
```

```python
# Reemplazar NaN por vacío
df = df.fillna("")
```

```python
valores_unicos_pais = df["País"].unique()
valores_unicos_pais
```

```python
valores_unicos_acomodacion = df["Acomodación"].unique()
valores_unicos_acomodacion
```

```python
valores_unicos_noches = df["Noches"].unique()
valores_unicos_noches
```

```python
# Extraer solo el número
df["Noches"] = df["Noches"].astype(str).str.extract("(\d+)").astype(int)
df.head()
```

```python
valores_unicos_Fecha_hospedaje = df["Fecha hospedaje"].unique()
valores_unicos_Fecha_hospedaje
```

```python
# Diccionario de meses en español a números
meses = {
    "enero": "01", "febrero": "02", "marzo": "03", "abril": "04",
    "mayo": "05", "junio": "06", "julio": "07", "agosto": "08",
    "septiembre": "09", "octubre": "10", "noviembre": "11", "diciembre": "12"
}

# Extraer mes y año en columnas separadas
df[["Mes", "Año"]] = df["Fecha hospedaje"].str.lower().str.extract(r"(\w+) de (\d{4})")

# Reemplazar el mes con su número
df["Mes"] = df["Mes"].map(meses)

# Construir la fecha en formato "AAAA/MM/DD"
df["Fecha hospedaje"] = df["Año"] + "/" + df["Mes"] + "/01"

# Eliminar columnas auxiliares
df = df.drop(columns=["Mes", "Año"])
```

```python
df.head()
```

```python
valores_unicos_Grupo_viaje = df["Grupo viaje"].unique()
valores_unicos_Grupo_viaje
```

```python
valores_unicos_Fecha_reseña = df["Fecha reseña"].unique()
valores_unicos_Fecha_reseña
```

```python
# Diccionario de meses en español a números
meses = {
    "enero": "01", "febrero": "02", "marzo": "03", "abril": "04",
    "mayo": "05", "junio": "06", "julio": "07", "agosto": "08",
    "septiembre": "09", "octubre": "10", "noviembre": "11", "diciembre": "12"
}

# Extraer día, mes y año en columnas separadas
df[["Día", "Mes", "Año"]] = df["Fecha reseña"].str.lower().str.extract(r"(\d{1,2}) de (\w+) de (\d{4})")

# Reemplazar el mes con su número
df["Mes"] = df["Mes"].map(meses)

# Formatear la fecha como "AAAA/MM/DD"
df["Fecha reseña"] = df["Año"] + "/" + df["Mes"] + "/" + df["Día"].str.zfill(2)  # Agregar ceros si es necesario

# Eliminar columnas auxiliares
df = df.drop(columns=["Día", "Mes", "Año"])
df.head()
```

```python
valores_unicos_Titulo = df["Titulo"].unique()
valores_unicos_Titulo
```

```python
valores_unicos_calificacion = df["Calificación"].unique()
valores_unicos_calificacion
```

```python
valores_unicos_resenna = df["reseña"].unique()
valores_unicos_resenna
```
